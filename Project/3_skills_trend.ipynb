```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
import seaborn as sns 


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_CAN = df[(df['job_title']== 'Data Analyst') & (df['job_country'] == 'Canada')].copy()
df_DA_CAN
```

```python
df_DA_CAN['job_posted_month_no'] = df_DA_CAN['job_posted_date'].dt.month
df_DA_CAN
```

```python
df_DA_CAN_explode = df_DA_CAN.explode('job_skills')
```

```python
df_DA_CAN_pivot = df_DA_CAN_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_CAN_pivot.loc['Total'] = df_DA_CAN_pivot.sum()
df_DA_CAN_pivot[df_DA_CAN_pivot.loc['Total'].sort_values(ascending=False).index]

```

```python
df_DA_CAN_pivot = df_DA_CAN_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_CAN_pivot.loc['Total'] = df_DA_CAN_pivot.sum()
df_DA_CAN_pivot = df_DA_CAN_pivot[df_DA_CAN_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_CAN_pivot = df_DA_CAN_pivot.drop('Total')
df_DA_CAN_pivot
```

```python
DA_totals = df_DA_CAN.groupby('job_posted_month_no').size()
type(DA_totals)
```

```python
df_DA_CAN_perc = df_DA_CAN_pivot.div(DA_totals/100, axis=0)
df_DA_CAN_perc
```

```python
df_DA_CAN_perc = df_DA_CAN_perc.reset_index()
df_DA_CAN_perc['job_posted_month'] = df_DA_CAN_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_CAN_perc = df_DA_CAN_perc.set_index('job_posted_month')
df_DA_CAN_perc = df_DA_CAN_perc.drop(columns='job_posted_month_no')
df_DA_CAN_perc
```

```python
df_plot = df_DA_CAN_perc.iloc[:, :5]

from matplotlib.ticker import PercentFormatter

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the Canada')
plt.ylabel('Skill Likelihood in job Posting')
plt.xlabel('')

ax= plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

plt.show()
```
